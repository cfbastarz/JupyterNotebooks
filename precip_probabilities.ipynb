## Plotando toda a série de precipitação:

```python
plt.plot([i[0] for i in all_precs],[i[1] for i in all_precs],marker="o")
plt.axhline(y=(np.mean([i[1] for i in all_precs])),color="red",label="Média")
plt.ylabel("Precipitação (mm)")
plt.xlabel("Anos")
plt.title("Série Temporal de Precipitação")
plt.legend()
plt.grid()
```

## Qual a probabilidade de ocorrência das precipitações observadas?

### Ordenando a série de precipitação (da maior para a menor):

```python
s_all_precs = sorted(all_precs, key=lambda tup: tup[1], reverse=True)
```

```python
print(s_all_precs)
```

### Calculando a probabilidade de ocorrência das precipitações a partir da série ordenada:

```python
o_s_all_precs = [(1989, 200,1),(1990, 140,2),(1980, 120,3),(1977, 110,4),(1982, 103,5),(1993, 95,6),(1981, 90,7),(1992, 90,7),(1984, 82,8),(1986, 75,9),(1978, 65,10),(1994, 64,11),(1988,60,12),(1991, 54,13),(1987, 45,14),(1985, 39,15),(1995, 19,16),(1979, 15,17),(1983, 10,18)]
```

```python
print(o_s_all_precs)
```

### Cálculo das Probabilidades (ver http://indico.ictp.it/event/a12165/session/21/contribution/16/material/0/0.pdf para mais informações):

```python
probs = [(i[2]/(len(o_s_all_precs)+1)*100) for i in o_s_all_precs]
```

```python
print(probs)
```

```python
o_s_precs = [i[1] for i in o_s_all_precs]
```

```python
plt.figure(figsize=(10,10))
plt.plot(probs,o_s_precs,marker="o")
plt.xticks(probs,fontsize=15)
plt.yticks(o_s_precs,fontsize=15)
plt.ylabel("Precipitação (mm)",fontsize=15)
plt.xlabel("Probabilidade (%)",fontsize=15)
plt.title("Probabilidade de Precipitação",fontsize=20)
plt.grid()
```
